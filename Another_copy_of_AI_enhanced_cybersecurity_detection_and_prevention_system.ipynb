```python
import pandas as pd

# Probe the number of columns to decide header set
probe_train = pd.read_csv("/content/KDDTrain+.txt", header=None, nrows=5)
n_cols = probe_train.shape[1]

cols_41 = [
    "duration","protocol_type","service","flag","src_bytes","dst_bytes",
    "land","wrong_fragment","urgent","hot","num_failed_logins","logged_in",
    "num_compromised","root_shell","su_attempted","num_root","num_file_creations",
    "num_shells","num_access_files","num_outbound_cmds","is_host_login",
    "is_guest_login","count","srv_count","serror_rate","srv_serror_rate",
    "rerror_rate","srv_rerror_rate","same_srv_rate","diff_srv_rate",
    "srv_diff_host_rate","dst_host_count","dst_host_srv_count","dst_host_same_srv_rate",
    "dst_host_diff_srv_rate","dst_host_same_src_port_rate","dst_host_srv_diff_host_rate",
    "dst_host_serror_rate","dst_host_srv_serror_rate","dst_host_rerror_rate",
    "dst_host_srv_rerror_rate","label"
]
cols_42 = cols_41 + ["difficulty"]
cols_43 = cols_42 # Assuming the extra column is the difficulty in both train and test

if n_cols == 41:
    use_cols = cols_41
elif n_cols == 42:
    use_cols = cols_42
elif n_cols == 43: # Handle the case with 43 columns
    use_cols = cols_43
else:
    raise ValueError(f"Unexpected number of columns: {n_cols}. Expected 41, 42 or 43.")

df_train = pd.read_csv("/content/KDDTrain+.txt", names=use_cols, header=None)
df_test  = pd.read_csv("/content/KDDTest+.txt"
,  names=use_cols, header=None)

print("✅ Loaded successfully.")
print("Train shape:", df_train.shape, "| Test shape:", df_test.shape)
print("\nFirst 5 rows (train):")
display(df_train.head())

print("\nColumn dtypes:")
print(df_train.dtypes.head(15))

print("\nMissing values (train):")
print(df_train.isnull().sum().sum())

print("\nLabel distribution (train):")
print(df_train['label'].value_counts())
```

```python
#STEP5-----preprocess&prepare features
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline

#1, Drop 'difficulty' if it exists
def drop_optional_cols(df):
  cols=df.columns.tolist()
  if "difficulty" in cols:
    df=df.drop(columns=["difficulty"])
  return df
df_train= drop_optional_cols(df_train)
df_test=drop_optional_cols(df_test)

#2, Define target(binary):benign vs attack
#  NSL-KDD uses 'normal' for benign; any other label is an attack.
df_train= df_train.copy()
df_test=df_test.copy()

df_train['target']=(df_train['label'] != 'normal').astype(int)
df_test['target']=(df_test['label'] != 'normal').astype(int)

#3, Separate features& target
X_full=df_train.drop(columns=['label','target'])
y_full=df_train['target'].values

X_test_external=df_test.drop(columns=['label','target'])
y_test_external=df_test['target'].values

#4,Identify categorical & numeric columns
categorical_cols=['protocol_type', 'service', 'flag']
numeric_cols=[c for c in X_full.columns if c not in categorical_cols]

#5, Preprocessing: OneHot for categorical, passthrough numeric
preprocess=ColumnTransformer(
    transformers=[
        ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False), categorical_cols)
    ],
    remainder='passthrough',      #keep numeric features
    verbose_feature_names_out=False  #cleaner feature names
)

#6, Train/validation split with stratify
X_train, X_val, y_train, y_val= train_test_split(
    X_full, y_full, test_size=0.2, random_state=42, stratify=y_full
)

print("Preprocessing setup complete.")
print("Train shapes: ", X_train.shape, y_train.shape)
print("Val shapes: ",X_val.shape, y_val.shape)
print("\nClass balance (train):")
unique, counts= np.unique(y_train, return_counts=True)
print(dict(zip(["benign(0)", "attack(1)"], counts)))
```

```python
# Step 6

from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score
import joblib

# 1) Build the end-to-end pipeline: preprocess -> classifier
rf = RandomForestClassifier(
    n_estimators=300,
    max_depth=None,
    n_jobs=-1,
    class_weight="balanced",
    random_state=42
)

rf_pipeline = Pipeline(steps=[
    ('preprocess', preprocess),
    ('clf', rf)
])

# 2) Fit
rf_pipeline.fit(X_train, y_train)

# 3) Evaluate on validation
y_val_pred = rf_pipeline.predict(X_val)
y_val_proba = rf_pipeline.predict_proba(X_val)[:, 1]  # prob of class 1 (attack)

print("🔎 Validation Accuracy:", round(accuracy_score(y_val, y_val_pred), 4))
print("\nValidation Classification Report:")
print(classification_report(y_val, y_val_pred, target_names=["benign", "attack"]))

# Some releases can have AUC issues if only one class appears in a small split; guard it:
try:
    print("Validation ROC-AUC:", round(roc_auc_score(y_val, y_val_proba), 4))
except Exception as e:
    print("Validation ROC-AUC: (could not compute) ->", e)

# 4) Evaluate on external test set (unseen KDDTest+)
y_test_pred = rf_pipeline.predict(X_test_external)
y_test_proba = rf_pipeline.predict_proba(X_test_external)[:, 1]

print("\n🧪 External Test Accuracy:", round(accuracy_score(y_test_external, y_test_pred), 4))
print("\nExternal Test Classification Report:")
print(classification_report(y_test_external, y_test_pred, target_names=["benign", "attack"]))

try:
    print("External Test ROC-AUC:", round(roc_auc_score(y_test_external, y_test_proba), 4))
except Exception as e:
    print("External Test ROC-AUC: (could not compute) ->", e)

# 5) Save the FULL pipeline (preprocess + model) — ready for FastAPI
MODEL_PATH = "/content/nslkdd_rf_pipeline.joblib"
joblib.dump(rf_pipeline, MODEL_PATH)
print(f"\n✅ Saved pipeline to: {MODEL_PATH}")
print("\nExternal TEst Classification Report: ")
print(classification_report(y_test_external, y_test_pred, target_names=["benign","attack"]))
try:
    pritn("external Test ROC-AUC:", round(roc_auc_score(y_test_external, y_test_proba), 4))
except Exception as e:
    print("external Test ROC-AUC: (couldnot compute) -> ", e)
MODEL
```
